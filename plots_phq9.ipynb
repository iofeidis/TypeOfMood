## Dynamics Features Ratios

$$ ratio = \frac{DynamicsFeature^{period_{2}}}{DynamicsFeature^{period_{1}}} $$

for example:

$$ ratio_{HoldTime.Mean()} = \frac{HoldTime.Mean()^{period_{2}}}{HoldTime.Mean()^{period_{1}}} $$

Absolute Difference:

$$ Absolute.Difference_{i} =  |{{DynamicsFeature^{period_{2}}} - {DynamicsFeature^{period_{1}}}} | $$

for example:

$$ Absolute.Difference_{HoldTime.Mean()} =  |{{HoldTime.Mean()^{period_{2}}} - {HoldTime.Mean()^{period_{1}}}} | $$
